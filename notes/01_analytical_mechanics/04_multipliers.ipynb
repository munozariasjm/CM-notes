#  Calculus of Variations & Lagrange Multipliers

It turns out that Euler Lagrange equation is acutally a simple way of optimizing generalized functionals. For instance if we have the following functional:

$$ J[y] = J(y(t))$$

where $y(t)$ is a function of $t$ (not necessarily the time) and $J$ is a functional of $y$. We can find the optimal $y$ by extremizing $J$:

$$
0 = \delta J = \int_{t_1}^{t_2} dt \left(\frac{\partial J}{\partial y} \delta y + \delta \dot{y} \frac{\partial J}{\partial \dot{y}}\right)
$$

Now we'll use integration by parts to get rid of the $\delta \dot{y}$ term:

$$
\delta J = \left(\frac{\partial J}{\partial \dot{y}} \delta y - \frac{d}{dt} \frac{\partial J}{\partial \dot{y}} \delta y\right)|_{t_1}^{t_2} + \int_{t_1}^{t_2} dt \left(\frac{\partial J}{\partial y} - \frac{d}{dt} \frac{\partial J}{\partial \dot{y}}\right) \delta y
$$

But since the variation at the endpoints is zero, we have:

$$
0 = \int_{t_1}^{t_2} dt \left(\frac{\partial J}{\partial y} - \frac{d}{dt} \frac{\partial J}{\partial \dot{y}}\right) \delta y \\ 
\therefore \frac{\partial J}{\partial y} - \frac{d}{dt} \frac{\partial J}{\partial \dot{y}} = 0
$$

## Largrange Multipliers

Lets now suppose all the forces are holonomic, and we have $k$ total constraits (labelled by $\alpha$). Plugging in D'Alembert's principle, we have:

$$\left[\frac{d}{dt}\frac{\partial L}{\partial \dot{q}_i} - \frac{\partial L}{\partial q_i} + \sum_{\alpha=1}^k \lambda_\alpha \frac{\partial f_\alpha}{\partial q_i} \right] \delta q_i = 0$$

It is actually safe to do so, because virtual work must be zero and lagrange multipliers $\lambda_\alpha$ are arbitrary.
